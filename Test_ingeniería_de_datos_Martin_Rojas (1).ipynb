```python
import pandas as pd
import requests
import json
from pandas import json_normalize
import re
```

**Leyendo archivo csv desde Git**

```python
tratados = pd.read_csv("https://raw.githubusercontent.com/sebas317/testDataEnginner/main/Tratados_internacionales_de_Colombia.csv")
```

**Realizando ajustes a los datos**

```python
tratados = tratados.rename(columns={'Estados-Organismos':'EstadosO'})
tratados.EstadosO = tratados.EstadosO.astype(str).str.upper()
trata2 = tratados.replace(to_replace='(NO REGISTRA)', value='NULL')
trata2['Fecha de Adopcion'] = pd.to_datetime(trata2['Fecha de Adopcion'], errors='coerce').astype(str)
trata2['Fecha Ley Aprobatoria'] = pd.to_datetime(trata2['Fecha Ley Aprobatoria'], errors='coerce').astype(str)
trata2['Sentencia Fecha Ley'] = pd.to_datetime(trata2['Sentencia Fecha Ley'], errors='coerce').astype(str)
trata2['Decreto Fecha Diario Oficial'] = pd.to_datetime(trata2['Decreto Fecha Diario Oficial'], errors='coerce').astype(str)
trata2.info()
```

**Convirtiendo en booleanos las dos columnas**

```python
trata2[['Bilateral', 'Vigente']] = trata2.apply(lambda row:  row[['Bilateral', 'Vigente']] == 'SI', axis=1).astype(bool) #time 47s
```

**Función lambda para contar numero de fronteras, e insertando datos**

```python

info_paises.insert(26, "Cantidad fronteras", info_paises.apply(lambda row:  int(len(row['borders']) if isinstance(row['borders'] , list)else ('0')), axis=1), allow_duplicates=False)
```

**Función para determinar diferencia horaria y función lambda para poder sacar los numeros de la lista convirtiendolos en enteros  para poderlos usar.**

```python
def d_hor(valor):
  if len(valor) == 2:
      valor[1]= valor[1]/100
      if valor[0]<0:
          num = (-5)+(valor[0]+valor[1])
      else:
          num = (-5)-(valor[0]+valor[1])
  else:
       if len(valor)== 0:
          return 'NaN'
       else:
           if valor[0]<0:
               num = (-5)+ (valor[0])
           else:
               num = (-5)-(valor[0])
  return "{:.2f}".format(num)





info_paises.insert(25, "Diferencia horas zona horaria", info_paises.apply(lambda row: d_hor(list(map(float, re.findall(r'-?\d+\.?\d*', str("".join(map(str, row['Zona horaria']))))))),axis=1), allow_duplicates=False)
info_paises.head(5)
```
